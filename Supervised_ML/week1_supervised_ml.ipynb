```python
import numpy as np
import matplotlib.pyplot as plt

#Some training data
X = np.array([1, 2, 3, 4, 5])

y = np.array([1.2, 1.9, 3.1, 3.9, 5.1])

plt.scatter(X, y)
plt.xlabel("Features")
plt.ylabel("labels")
plt.title("training data")
plt.grid(True)
plt.show()
```

```python
#Linear Model
def predict(x, w, b):
    return w * x + b
```

```python
#Cost Function
def compute_cost(X, y, w, b):
    m = len(X)
    cost = 0

    for i in range(m):
        f_wb = predict(X[i], w, b)
        cost += (f_wb - y[i]) ** 2

        return cost / (2*m)
```

```python
#Initial parameters
w, b = 0, 0
cost = compute_cost(X, y, w, b)
print(f"Initial cost: {cost}")
```

```python
#Gradient Descent

def compute_gradient(X, y, w, b):
    m = len(X)
    dj_dw = 0
    dj_db = 0

    for i in range(m):
        f_wb = predict(X[i], w, b)
        dj_dw = (f_wb - y[i]) * X[i]
        dj_db = (f_wb - y[i])
        
    return dj_dw/m, dj_db/m
```

```python
iterations = 1000
alpha = 0.01
w , b = 0.0, 0.0
cost_history = []

print("Iteration   | Cost                   | w                  | b                   ")
for i in range(iterations):
    dj_dw, dj_db = compute_gradient(X, y, w, b)
    w -= alpha * dj_dw
    b -= alpha * dj_db

    if i % 100 == 0:
        cost = compute_cost(X, y, w, b)
        cost_history.append(cost)
        
        print(f"{i}         | {cost}  | {w} | {b}")
```

```python
plt.scatter(X, y, label="Data")
plt.plot(X, predict(X, w, b), color="red", label="Prediction")
plt.xlabel(X)
plt.ylabel(y)
plt.legend()
plt.title("Final Model Fit")
plt.grid(True)
plt.show()
```
